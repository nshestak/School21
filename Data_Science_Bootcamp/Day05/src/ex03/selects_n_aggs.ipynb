## 1. Загрузите файл JSON, который вы создали в предыдущем упражнении, во фрейм данных

## 2. Выполните следующие действия

- Отображать только те строки, в которых сумма Refund превышает 2100; 

```python
df[df['Fines'] > 2100]
```

- Отобразить только строки, где Fines больше чем 2100 и Refund равна 2; 

```python
df[(df['Fines'] > 2100) & (df['Refund'] == 2)]
```

- Отображать строки только в тех случаях, когда модели указаны в списке:  [’Focus’, ’Corolla’];

```python
# df[(df['Model'] == 'Focus') | (df['Model'] == 'Corolla')]
df[df['Model'].isin(['Focus', 'Corolla'])]
```

- Oтображать только те строки, в которых указан номер автомобиля из списка: [’Y7689C197RUS’, ’92928M178RUS’, ’7788KT197RUS’, ’H115YO163RUS’, ’X758HY197RUS’];

## 3. Cоздайте агрегированные данные с указанием марки и модели

- Отобразите median Fines, сгруппированные по Make;

```python
df.groupby('Make').median('Fines')
```

- Отобразите median Fines, сгруппированные по Make и Model; 

```python
df.groupby(['Make', 'Model']).median('Fines')
```

- Отобразите число Fines сгруппировать по Make и Model для того, чтобы понять, можно ли доверять meadian значениям;

```python
df.groupby(['Make', 'Model'])['Fines'].count()
```

- Отобразите минимальные и максимальные Fines, сгруппированные по Make и Model, чтобы лучше понять разницу; 

```python
df.groupby(['Make', 'Model'])['Fines'].agg(['min', 'max'])
```

- Отобразите стандартное отклонение Fines, сгруппированных по Make и Model, чтобы лучше понять разницу; 

```python
df.groupby(['Make', 'Model'])['Fines'].std()
```

## 4. Произведите агрегацию с указанием номера автомобиля 

- Oтобразите номера автомобилей, сгруппированные по количеству Fines в порядке убывания мы хотим найти тех, кто чаще всего нарушал закон; 

```python
df.groupby(['CarNumber'])['Fines'].count().sort_values(ascending=False)
```

- Выделите из исходного фрейма данных все строки, соответствующие номеру автомобиля топ-1, мы хотим немного увеличить масштаб;

```python
# top_car_num = df.index.value_counts().idxmax()
# df[df.index == top_car_num]
fines_count = df.groupby(['CarNumber'])['Fines'].count().sort_values(ascending=False)
df[df.index == fines_count.idxmax()]
```

- Отобразите номера автомобилей, сгруппированные по сумме Fines в порядке убывания, мы хотим найти тех, кто заплатил больше всего;

```python
df.groupby(['CarNumber'])['Fines'].sum().sort_values(ascending=False)
```

- Выделите из исходного фрейма данных все строки, соответствующие номеру автомобиля топ-1, мы хотим немного увеличить масштаб;

```python
fines_sum = df.groupby(['CarNumber'])['Fines'].sum().sort_values(ascending=False)
df[df.index == fines_sum.idxmax()]
```
