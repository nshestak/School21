```python
import pandas as pd
```

## 1. Создайте представления фрейма данных с двумя столбцами: datetime и user, прочитав feed-views.log

```python
df = pd.read_csv('../data/feed-views.log', sep='\t', header=None, names=['datetime', 'user'])
df
```

- Преобразуйте datetime в Dtype datetime64[ns];

```python
df['datetime'] = pd.to_datetime(df['datetime'])
df.info()
```

- Извлеките год, месяц, день, час, минуту и секунду из значений этого столбца в новые столбцы; 


```python
df_copy = df.copy()
```

```python
columns = ['year', 'month', 'day', 'hour', 'minute', 'second']

for column in columns:
    df_copy[column] = getattr(df_copy['datetime'].dt, column)
df_copy
```

## 2. Содайте новый столбец **daytime**

- Вам необходимо присвоить конкретное значение времени суток, если час находится в пределах определенного интервала, например, после полудня, если час больше 11 и меньше или равен 17;
- 0 – 3.59 ночи, 4 - 6.59 раннего утра, 7 – 10.59 утра, 11 – 16.59 дня, 17 – 19.59 раннего вечера, 20 – 23.59 вечера;
- Ипользуйте метод **cut** для решения этой подзадачи;

```python
df_copy['daytime'] = pd.cut(df_copy['hour'], bins=[-1, 3.99, 6.99, 10.99, 16.99, 19.99, 23.99], labels=['night', 'early morning', 'morning', 'afternoon', 'early evening', 'evening'])
df_copy
```

- Назначьте столбец **user** в качестве индекса; 

```python
df_copy.set_index('user', inplace=True)
df_copy
```

## 3. Вычислите количество элементов в вашем фрейме данных

- Используйте метод **count()**; 

```python
df_copy.count()
```

- Pассчитайте количество элементов в каждой категории 'daytime', используя метод **value_counts()**;

## 4. Сортируйте значения в вашем фрейме данных по часам, минутам и секундам в порядке возрастания (одновременно, а не по одному) 

```python
df_copy = df_copy.sort_values(by=['hour', 'minute', 'second'])
df_copy
```

## 5. Рассчитайте минимальное и максимальное значение для часов и моду для 'daytime'

- Рассчитайте максимальное время в час для строк, где время суток равно **night**;

```python
night_hours = df_copy[df_copy['daytime'] == 'night']
max_night_hour = night_hours['hour'].max()
max_night_hour
```

- Рассчитайте минимальное время в час для строк, где время суток - **morning**;

```python
morning_hours = df_copy[df_copy['daytime'] == 'morning']
min_morning_hour = morning_hours['hour'].min()
min_morning_hour
```

- Помимо этого узнайте, кто заходил на страницу в эти часы (приведите из этого один пример);

```python
df_copy[df_copy['hour'] == max_night_hour]
```

```python
df_copy[df_copy['hour'] == min_morning_hour]
```

- Рассчитать моду для hour и daytime;

```python
mode_hour = df_copy['hour'].mode()
mode_daytime = df_copy['daytime'].mode()
print(f'hour: {mode_hour[0]}, daytime: {mode_daytime[0]}')
```

## 6. Отобразите 3 самых ранних часа morning и соответствующие имена пользователей, а также 3 последних часа и имена пользователей с помощью nsmallest() и nlargest() 

```python
df_copy[df_copy['daytime'] == 'morning'].nsmallest(3, 'hour')
```
