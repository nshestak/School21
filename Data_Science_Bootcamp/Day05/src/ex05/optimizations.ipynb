## 2. Итерации: во всех следующих подзадачах вам необходимо рассчитать штрафы / возмещение * год для каждой строки и создать новый столбец с рассчитанными данными и измерить время используя волшебную команду %%timeit в ячейке

- Цикл: напишите функцию, которая выполняет итерацию по фрейму данных, используя for i in range(0, len(df)), iloc и append() для списка, присваивая результат выполнения
функции новому столбцу в фрейме данных;

```python
def cal_filnes(data):
    result = []
    for i in range(0, len(data)):
        result.append(data.iloc[i]['Fines'] / data.iloc[i]['Refund'] * data.iloc[i]['Year'])
    return result
```

```python
%%timeit
df['CalFines'] = cal_filnes(df)
```

```python
df
```

- Сделайте это с помощью iterrows();

```python
def cal_filnes(data):
    result = []
    for index, row in data.iterrows():
        result.append(row['Fines'] / row['Refund'] * row['Year'])
    return result
```

```python
%%timeit
df['CalFines'] = cal_filnes(df)
```

```python
df
```

- Cделайте это с помощью apply() и лямбда-функции;

```python
def cal_filnes(data):
    result = data.apply(lambda x: x['Fines'] / x['Refund'] * x['Year'], axis=1)
    return result
```

```python
%%timeit
df['CalFines'] = cal_filnes(df)
```

```python
df
```

- Cделайте это, как в предыдущей подзадаче, но с помощью метода .values;

```python
def cal_filnes(data):
    result = data['Fines'].values / data['Refund'].values * data['Year'].values
    return result
```

```python
%%timeit
df['CalFines'] = cal_filnes(df)
```
