```python
from sympy import *
i,n = symbols('i n')

# перебирает элементы i от 1 до n,
# затем умножает и суммирует

summation = Sum(2*i,(i,1,n))

# задает n равным 5,
# перебирает числа от 1 до 5

up_to_5 = summation.subs(n, 5)
print(up_to_5.doit()) # 30
```
